### Importing Necessary Libraries

```python
import pandas as pd
import numpy as np
import re
```

#### Importing dataset 

```python
df = pd.read_csv("Literature_&_Fiction_Books_2022.csv")
```

```python
unclean_data = df.copy() 
```

#### Checking for Null values in each Column

```python
print(unclean_data.isna().sum().sort_values(ascending = False))
```

### Cleaning Release_Date column 
- Removing redundancy and white spaces

```python
unclean_data['Release_Date'] = unclean_data['Release_Date'].str.lstrip('Release Date:\n')
```

```python
unclean_data.head(5)
```

### Cleaning "Rating" and "No_of_Ratings" columns
- Renaming "Rating" column
- Removing redundancy 
- Coverting values into float data type

```python
unclean_data['Score'] = unclean_data['Rating'].str.replace(r'\ out of 5 stars$', '', regex=True).fillna(0).astype('float32')

unclean_data['No_of_Ratings'] = unclean_data['No_of_Ratings'].replace('Not rated yet', 0) #Replacing "Not rated yet" with zero
#Scraping off text converting numbers into int32 data type
unclean_data['No_of_Ratings'] = unclean_data['No_of_Ratings'].replace(to_replace='[^0-9]', regex=True, value='').astype('int32') 

unclean_data.head()
```

### Cleaning "Regular_Price" column
- Renaming "Regular_Price" column
- Removing redundancy 
- Coverting values into float data type

```python
unclean_data['Price'] = unclean_data['Regular_Price'].replace(".00$", '', regex=True).fillna(0)
unclean_data['Price(INR)'] = unclean_data['Price'].replace(to_replace='[^0-9]', regex=True, value='').fillna(0).astype('float32')
unclean_data.head()
```

### Cleaning "Language" column
- Removing redundancy 

```python
unclean_data['Language'] = unclean_data['Language'].replace(to_replace='[\W_]+', regex=True, value='')
unclean_data['Language'] = unclean_data['Language'].replace(to_replace='Language', regex=True, value='')
unclean_data.head()
```

### Cleaning "Book_Audio_Length" column
- Renaming "Book_Audio_Length" column
- Extracting hours and minutes into two seperate columns in a temporary data frame
- Converting them into int32 data type
- Concatenating them into hh:mm:ss format with panda's to_timedelta

```python
time_name = {0:"hours",1:"minutes"}
temp_df = unclean_data['Book_Audio_Length'].str.extract(r'(\d+)[^\d]+(\d+)').fillna(0).astype('int32')
temp_df.rename(columns=time_name,
              inplace=True)
```

```python
temp_df = pd.to_timedelta(temp_df['hours'].astype(str) + ':' + temp_df['minutes'].astype(str)+":00")
temp_df = temp_df.astype(str).map(lambda x: x[7:]) #Striping off days values which comes along with pd.to_timedelta
```

```python
temp_df = temp_df.to_frame()
temp_df.rename( columns={0 :'Audio_Book_Length'}, inplace=True )
result = unclean_data.join(temp_df)
```

### Dropping old columns

```python
result.drop(['Rating','Regular_Price','Book_Audio_Length','Price'], axis=1, inplace = True)
```
